```python
import pandas as pd
people =["Kiran","Arjun","Vijay","Varun"];
age=[25,30,35,40]
height=[145,151,165,173]
weight=[45,55,65,75]
data={
    "Name": people,
    "Age":age,
    "Height":height,
    "Weight":weight
}
df=pd.DataFrame(data)
print(df)



```

```python
# 3. Scatter Plot: Age vs Weight
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Weight'], color='blue', marker='o')
plt.title('Scatter Plot of Age vs Weight')
plt.xlabel('Age (years)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
# 4. Bar Chart: Height of Individuals
plt.figure(figsize=(8, 6))
plt.bar(df['Name'], df['Height'], color='green')
plt.title('Bar Chart of Individuals\' Heights')
plt.xlabel('Name')
plt.ylabel('Height (cm)')
plt.show()
```

```python
# 5. Histogram: Distribution of Age
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```
